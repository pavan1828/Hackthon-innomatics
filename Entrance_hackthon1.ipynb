```python
data = df2.groupby('movieId').agg({'rating': ['count', 'mean']}).reset_index()
data.columns = ['movieId', 'rating_count', 'rating_mean']

x = pd.merge(df1, data, on='movieId', how='inner')
y = x[x['rating_count'] > 50]
y
```
